### Several ways of finding elements in BS4

https://youtu.be/gRLHr664tXA

```python
import requests
from bs4 import BeautifulSoup
import re

url = 'https://www.newegg.com/gigabyte-geforce-rtx-3070-gv-n3070vision-oc-8gd/p/N82E16814932450'

headers =  {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.63 Safari/537.36'}
r = requests.get(url, headers=headers)

print(r.status_code)
#print(r.text)

soup = BeautifulSoup(r.text, 'html.parser')
#print(soup.prettify())

# find first element with tag name
elements = soup.find('a')
print(elements)

# find all elements with tag name
elements = soup.find_all('a')
print(len(elements))
print(elements[0])

# find text content
dollar = soup.find(text='$')
# get it's element
element = dollar.parent
print(element)
# get element with 'strong' tag under element
strong = element.find('strong')
print(strong)
# get price which is it's text
print(strong.text)
```

```python
url = 'https://raw.githubusercontent.com/techwithtim/Beautiful-Soup-Tutorial/main/index2.html'
r = requests.get(url, headers=headers)

print(r.status_code)
#print(r.text)

soup = BeautifulSoup(r.text, 'html.parser')
#print(soup.prettify())

option = soup.find('option')
print(option)

# change value in soup
option['value'] = 'blah-type'
print(option)

# add attribute
option['color'] = 'red'
print(option)

# get all attributes as dict
print(option.attrs)

# get all elements that match the tags in a list
elements = soup.find_all(['p', 'div', 'li'])
print(elements)

# find element with tag and text
element = soup.find(['option'], text = 'Undergraduate')
print(element)

# find element with tag and attribute
element = soup.find(['option'], color = 'red')
print(element)

# find elements of specific class
# use class_ param because class is used by python
buttons = soup.find_all(class_ = 'btn-item')
print(buttons)

# use regular expression. Anything that start with $
dollars = soup.find_all(text = re.compile('\$.*'))
print(dollars)

# limit gives a maximum elements to find
dollars = soup.find_all(text = re.compile('\$.*'), limit = 1)
print(dollars)

# find all input tag elements with type 'text'
# change placeholder attribute
inputs =  soup.find_all('input', type = 'text')
for element in inputs:
    element['placeholder'] = 'I changed you!'

with open('changed.html', 'w') as file:
    file.write(str(soup))
```

```python
url = 'https://coinmarketcap.com'

r = requests.get(url, headers=headers)

print(r.status_code)
#print(r.text)

soup = BeautifulSoup(r.text, 'html.parser')
#print(soup.prettify())

# path searching directly. fo to element with tbody tag
tbody = soup.tbody
#print(tbody.prettify())

# all tags inside tbody
trs = tbody.contents
print(len(trs))

# siblings
a = trs[0]
b = a.next_sibling
c = b.previous_sibling
print(a.text)
print(b.text)
print(c.text)

# all siblings after b
rest = b.next_siblings
print(len(list(rest)))

# get parent and print tag name
print(c.parent.name)

# contents gives a list and children gives a generator

# recursive iteration through all descendants, generator
print(len(list(tbody.descendants)))

# get prices crypto
prices = {}

for tr in trs:
    # get 3rd and 4th element of coontents
    name, price =  tr.contents[2:4]
    prices[name.text] = price.text

print(prices)
```

```python
# searching newegg fo '3080' and 'In stock'
url = 'https://www.newegg.com/p/pl?d=3080&N=4131'

r = requests.get(url, headers=headers)

print(r.status_code)
#print(r.text)

soup = BeautifulSoup(r.text, 'html.parser')
#print(soup.prettify())

# find amount of pages with search result
page = soup.find(class_ = 'list-tool-pagination-text').find('strong')
pages = int(page.contents[4])
print(pages)

found = {}
for page in range(pages):
    #print('\nPAGE %s:' % (page+1))
    url = 'https://www.newegg.com/p/pl?d=3080&N=4131&page=%s' % (page+1)
    r = requests.get(url, headers=headers)
    if r.status_code != 200: break

    soup = BeautifulSoup(r.text, 'html.parser')

    for itemGrid in soup.find_all('div', class_ = 'items-grid-view'):
        # check if '3080' in text
        for item in itemGrid.find_all(class_ = 'item-title', text = re.compile('3080')):
            found[item.text] = {}
            found[item.text]['link'] = item['href']
            # traveres all up parents to find the on with this class
            container = item.find_parent(class_ = 'item-container')
            price = container.find(class_ = 'price-current')
            if price.text != '':
                found[item.text]['price'] = float(price.text.split()[0].lstrip('$').replace(',', ''))
            else:
                found[item.text]['price'] = 0.0

sorted_items = sorted(found.items(), key = lambda x: x[1]['price'])

for item in sorted_items:
    print(item[0])
    print(item[1]['price'])
    print(item[1]['link'])
    print('-----------------------')
```
