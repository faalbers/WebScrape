```python
tickers = ['FB', 'AMZN', 'NFLX', 'GOOG']
```

### Options data

```python
tickers = [yf.Ticker(ticker) for ticker in tickers ]
tickers
```

```python
dfs = []
for ticker in tickers:
    # get financial statement for each ticker
    pnl = ticker.financials
    bs = ticker.balancesheet
    cf = ticker.cashflow
    fs = pd.concat([pnl,bs,cf])

    # make data frame foormat nicer
    data = fs.T
    data = data.reset_index() # reset index (data) into column
    data.columns = ['Date', *data.columns[1:]] # rename first column with 'Date'
    data['Ticker'] = ticker.ticker # add ticker name
    dfs.append(data)
data
```
